```python
a = ["props"]*7
# print(a)

import numpy as np

arr = np.reshape(np.asarray(np.arange(0, 9, step=1)), (3,3))
# print(arr)

import pandas
df = pandas.DataFrame(arr, columns=[['col1', 'col2', 'col3']])
# print(df)
a = np.asarray([df, 'string'], dtype=object)
# print(a[0])
# df['col1']
# for i in df:
#     print(df[i])

arr1 = arr[:1]
# print(arr1)

# print(range(1))

time_array = np.arange(start=0, stop=2, step=1)
# print('0 = ', time_array[0])
# print(time_array[-1])
b=['t'+str(i) for i in np.arange(0,time_array[-1])]

# print(b)
# import pyvista
# import numpy as np
# mesh = pyvista.Sphere()
# mesh.add_field_data(np.arange(10), 'my-field-data')
# mesh['my-field-data']
# mesh.plot()

# x,y,z = np.mgrid[:3, :3, :3]
# print('x = \n', x)
# print('y = \n', y)
# print('z = \n', z)

# from pyvista import examples

# mesh = examples.load_hexbeam()

# from pyvistaqt import BackgroundPlotter

# pl = BackgroundPlotter()
# pl.add_mesh(mesh)

# mesh.points[:, -1] *= 0.5

import pyvista as pv

import numpy as np

grid = pv.Wavelet()
outline = grid.outline()

def createGrid(tstep):
    """Helper to yeild a timestep mesh in place of missing code."""
    j = 10
    return pv.Wavelet(x_freq=tstep+j, y_freq=tstep+j, z_freq=tstep+j)


class SceneEngine:
    def __init__(self, grid, ):
        self.grid = grid
        self.output = grid.threshold(all_scalars=False)
        
        self._mi = 0
        self._ma = 1
        self._tstep = 0
    
    def _threshold(self):
        self.output.overwrite(self.grid.threshold([self._mi, self._ma], all_scalars=False))
        
    def threshold_min(self, mi):
        self._mi = mi
        self._threshold()
        
    def threshold_max(self, ma):
        self._ma = ma
        self._threshold()
        
    def update_timestep(self, tstep):
        self._tstep = tstep
        self.grid.overwrite(createGrid(tstep))
        self._threshold()

e = SceneEngine(grid)


p = pv.Plotter(notebook=False)
# add outline
p.add_mesh(outline, color="k")
p.add_mesh(e.output, cmap='viridis', opacity=1)
    
# add results data
p.add_slider_widget(e.update_timestep, rng=[0, 400], value=0, title='Timestep', 
                    pointa=(.67, .9), pointb=(.98, .9), event_type='always')
p.add_slider_widget(e.threshold_min, rng=[np.nanmin(grid.active_scalars), np.nanmax(grid.active_scalars)],
    value=np.nanmin(grid.active_scalars), title="Threshold Min",
    pointa=(.025, .9), pointb=(.31, .9)
    )
p.add_slider_widget(e.threshold_max, rng=[np.nanmin(grid.active_scalars), np.nanmax(grid.active_scalars)],
    value = np.nanmax(grid.active_scalars), title="Threshold Max",
    pointa=(.35, .9), pointb=(.64, .9)
)

p.show()

```
